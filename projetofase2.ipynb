```python

import pandas as pd
import streamlit as st

# 1. Carrega o CSV gerado
df_pacientes = pd.read_csv('pacientes_df.csv')

# 2. Conta a quantidade de pacientes por Região Administrativa
contagem_por_regiao = df_pacientes['regiao_administrativa'].value_counts()

# Mostra no console (Terminal)
print(contagem_por_regiao)

# Mostra no Streamlit (Interface Web)
st.bar_chart(contagem_por_regiao)
```
